# Data Preparation and Analysis

One of the most important aspects of deploying a ML solution is the preparation and analysis of the data set.
Once you understand your data, you will make a more informed decision about what types of ML algorithms apply.

This example illustrates a likely partitioning and analysis process using the Iris data set.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams.update({'font.size': 18})

from sklearn import datasets
```

The Iris dataset is a collection of 150 samples of Iris flower features and **human made** classifications.
(You can see more detail [here](https://en.wikipedia.org/wiki/Iris_flower_data_set))

```python
iris_dataset = datasets.load_iris()
X = iris_dataset.data
print("Features ({1}): \n{0}".format(iris_dataset.feature_names, len(iris_dataset.feature_names)))
```

```python
print("Data set shape: {0}".format(X.shape))
print("Number of instances: {0}".format(X.shape[0]))
```

```python
y = iris_dataset.target
print("Classes ({0}): \n{1}".format(len(iris_dataset.target_names), iris_dataset.target_names))
```

Before we analyze, we need to **partition** the data into **training** and **test** data.
Thankfully, sklearn automates this process easily with the `train_test_split` function.
All of our analysis will be performed on the training set.
Do not cheat and use the test data!

```python
from sklearn.model_selection import train_test_split
# Save 20% of the data for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

Histograms are useful to get a quick glance at each feature's value **distribution**.
To select the number of bins, a good rule of thumb is Sturge's rule:

$$
N = \lceil log_2(M) \rceil + 1
$$

where $M$ is the number of data instances.

```python
from math import log2, ceil

n_bins = ceil(log2(X_train.shape[0]))
print("Number of bins = {0}".format(n_bins))
```

```python
fig1 = plt.figure(num=1,figsize=(8,5))
plt.subplot(221)
plt.xlabel(iris_dataset.feature_names[0])
plt.hist(X_train[:,0], bins=n_bins)
plt.subplot(222)
plt.xlabel(iris_dataset.feature_names[1])
plt.hist(X_train[:,1], bins=n_bins)
plt.show()
```

```python
fig2 = plt.figure(num=2, figsize=(8,5))
plt.subplot(221)
plt.xlabel(iris_dataset.feature_names[2])
plt.hist(X_train[:,2], bins=n_bins)
plt.subplot(222)
plt.xlabel(iris_dataset.feature_names[3])
plt.hist(X_train[:,3], bins=n_bins)
plt.show()
```

Scatter plots are also very useful to visually search for correlations in the data.
The following plots will fix the **sepal length** and compare it to the other three.

```python
plot_idxs = [1,2,3]
fig3 = plt.figure(num=3, figsize=(8,10))
classes = iris_dataset.target_names
for idx in plot_idxs:
    plt.subplot(3,1,idx)
    plt.scatter(X_train[:,0], X_train[:,idx], c=y_train)
    plt.ylabel(iris_dataset.feature_names[idx])
plt.xlabel(iris_dataset.feature_names[0])
plt.show()
```

Interesting! There does appear to be a pattern forming in this data: the setosa class usually has a smaller sepal length compared to the other features.
Also, this class appears to be **linearly separable** from the rest of the other two classes.
This insight can shape how we select our machine learning algorithm.

As discussed in class, tSNE is another nice visualization tool that builds clusters of related data in the data set and embeds it into a 2D or 3D space.

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X_train)
```

```python
fig4 = plt.figure(num=4)
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y_train)
plt.show()
```
